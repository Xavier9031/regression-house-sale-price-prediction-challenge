# Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# Setting

```python
# Data Path
path_train = os.path.join(os.getcwd(), "data", "train-v3.csv")
path_valid = os.path.join(os.getcwd(), "data", "valid-v3.csv")
path_test = os.path.join(os.getcwd(), "data", "test-v3.csv")
path_meta = os.path.join(os.getcwd(), "data", "metadata.csv")
```

# Data Load

```python
# Load data
df_train = pd.read_csv(path_train)
df_valid = pd.read_csv(path_valid)
df_test = pd.read_csv(path_test)
df_meta = pd.read_csv(path_meta)

df_train_all = pd.concat([df_train, df_valid], axis=0).reset_index(drop=True)
```

```python
# check meta data
df_meta
```

```python
# check train data
df_train_all.info()
```

```python
# check target distribution
print(df_train_all['price'].describe())
plt.figure(figsize=(9, 8))
sns.histplot(df_train_all['price'], bins=100, kde=True)
```

```python
# check correlation
plt.figure(figsize=(10,8))
sns.heatmap(df_train_all.corr(), cmap="RdBu")
plt.title("Correlations Between Variables", size=15)
plt.show()
```
